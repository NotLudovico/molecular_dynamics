```python
# solution 1.
cell, positions = simplemd.generate_lattice(3)
print(positions.shape)
print(cell)
```

```python
# solution 2.
# reasonable input options
# note: initial coordinates are still missing
keys={
 'temperature': 0.722,
 'tstep': 0.005,
 'forcecutoff': 2.5,
 'listcutoff': 3.0,
 'nstep': 2000,
 'nconfig': 10,
 'nstat': 10
 }
# generate a crystal lattice and pass it:
keys["cell"] = cell
keys["positions"] = positions
```

```python
# solution 3.
# create runner
smd=simplemd.SimpleMD(**keys)
# run
smd.run()
```

```python
# solution 4.
# plotting instantaneous temperature
fig, ax = plt.subplots()
ax.plot(np.array(smd.statistics).T[2])
ax.axhline(y=keys["temperature"], color='r', linestyle='--')
ax.set_xlabel('step')
ax.set_ylabel('temperature')
plt.show()

# plotting potential energy
fig, ax = plt.subplots()
ax.plot(np.array(smd.statistics).T[3], label="potential energy")
ax.plot(np.array(smd.statistics).T[4], label="total energy")
ax.legend()
ax.set_xlabel('step')
ax.set_ylabel('energy')
plt.show()

# Temperature <=> Kinetic energy, drops down because particles are placed in minima of potential
# Energy stabilizes around "half" of total energy due to equipartition theorem
```

```python
# solution step 3.1
keys = {
    'temperature': 1.3,
    'friction': 1.0,
    'tstep': 0.005,
    'forcecutoff': 2.5,
    'listcutoff': 3.0,
    'nstep':10000,
    'nconfig': 10,
    'nstat': 10
    }
cell, positions = simplemd.generate_lattice(3)
keys["cell"] = cell
keys["positions"] = positions
smd = simplemd.SimpleMD(**keys)
smd.run()
# Periodic boundary condition in the sense that forces are computed with periodic boundary condition but not positions. 
# So this simulates a liquid
```

```python
# solution step 3.2
simplemd.write_trajectory(path+"/trajectory.xyz", smd.trajectory)
```

```python
#needs to be processed as text
with open(path+"/trajectory.xyz") as f:
    traj_xyz = f.read()

view = py3Dmol.view(width=400, height=300)
view.addModelsAsFrames(traj_xyz, "xyz")
view.setStyle({"sphere": {"radius":0.4}})
view.animate({'loop': "forward"})
view.zoomTo()
view.show()
```

```python
import numpy as np
total_time = 10

dts = [0.0001, 0.001, 0.005]
temps = [0.722, 0.1, 1, 2]
for temp in temps:
    for dt in dts:
        keys = {
            'temperature': temp,
            'tstep': dt,
            'forcecutoff': 2.5,
            'listcutoff': 3.0,
            'nstep':int(total_time / dt),
            'nconfig': 10,
            'nstat': 10
            }
        cell, positions = simplemd.generate_lattice(3)
        keys["cell"] = cell
        keys["positions"] = positions
        smd = simplemd.SimpleMD(**keys)
        smd.run()

        stat_name = f"./stat/{dt}-{temp}-stat.txt"
        np.savetxt(stat_name, smd.statistics)

        # name = f"./traj/{dt}-{temp}.xyz"
        # simplemd.write_trajectory(name, smd.trajectory)
```

```python
for temp in temps:
    fig, ax = plt.subplots()
    for dt in dts:
        stat_name = f"./stat/{dt}-{temp}-stat.txt"
        vals = np.genfromtxt(stat_name)
        ax.plot(np.arange(0, int(1 / dt) , 1) * (dt / 0.001),np.array(vals).T[4], label=f"{dt}-{temp}")
    ax.legend()
    
```

```python
# Fixed dt

for temp in temps:
    fig, ax = plt.subplots()
    stat_name = f"./stat/0.001-{temp}-stat.txt"
    vals = np.genfromtxt(stat_name)
    ax.plot(np.array(vals).T[3], label="potential energy")
    ax.plot(np.array(vals).T[4], label="total energy")
    title = f"Temp: {temp}"
    ax.set_title(title)
```
